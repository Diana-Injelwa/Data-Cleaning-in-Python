Avg Monthly Long Distance Charges:	Indicates the customer’s average long distance charges, calculated to the end of the quarter specified above (if the customer is not subscribed to home phone service, this will be 0)

```python
# Returns null values from 'Avg Monthly Long Distance Charges' and the corresponding values from 'Phone Service' column
df.loc[df['Avg Monthly Long Distance Charges'].isnull(), ['Avg Monthly Long Distance Charges', 'Phone Service']]
```
